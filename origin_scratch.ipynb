```python
#the leading suggestion is run this under a for loop
# setup
domain = 'https://python.zgulde.net'
endpoint = '/api/v1/items'
items = []

# For each page -- until next page is None
while endpoint is not None:
    #wow this was easier then I had orignially thought
    url = domain + endpoint
    response = requests.get(url)
    data = response.json()
    items.extend(data['payload']['items'])
    #  update the endpoint
    endpoint = data['payload']['next_page']
    # print(endpoint)
    print(items)
    print(endpoint)
    # print(pd.DataFrame(data['payload']))
pd.DataFrame(items)
```
